```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data
cars_df = pd.read_csv('cars93.csv')
#print(cars_df.columns)
```

### 1. Boxplot

```python
# Boxplot for revs per mile
plot_df=cars_df[['Manufacturer','Rev.per.mile']] # Select the relevant columns in preparation for plotting
manufacturer_list = ['Audi','Hyundai','Suzuki','Toyota']
plot_df=plot_df[(plot_df['Manufacturer'].isin(manufacturer_list))] # Select the relevant manufacturers

# Plot Boxplot
plt.figure()
sns.boxplot(data=plot_df, x='Rev.per.mile', y='Manufacturer')
plt.xlabel('Revs per mile')
plt.ylabel('Manufacturer')
plt.show()
plt.close()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:** Toyota


### 2. Histogram

```python
# Histogram of MPG
plot_df=cars_df[['MPG.city','MPG.highway']] # Select the data that I wish to plot

plt.figure()
sns.histplot(data=plot_df)
plt.xlabel('MPG')
plt.ylabel('Number of car models')
plt.show()
plt.close()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** Highway - The majority of car models have a higher MPG on the highway and so are more fuel efficient whilst driving there.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plot_df=cars_df[['Wheelbase','Turn.circle']] # select the data that I wish to make a lineplot for

plt.figure()
sns.lineplot(data=plot_df, x='Wheelbase', y = 'Turn.circle' )
plt.xlabel('Wheelbase')
plt.ylabel('Turning cirlce')
plt.show()
plt.close()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  There is a positive correlation between wheelbase and turning circle. As the wheelbase gets larger the turning circle gets larger.


### 4. Barplot

```python
# Barplot for mean horsepower
plot_df=cars_df.groupby('Type')['Horsepower'].mean() # Calculate the mean for each car type
size_order=['Small', 'Compact', 'Sporty', 'Midsize', 'Large', 'Van']  # Here I am estimating the size order of the various categories

plt.figure()
sns.barplot(data=plot_df, order=size_order)
plt.xlabel('Type')
plt.ylabel('Horsepower')
plt.show()
plt.close()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** Yes, generally a larger car means more horsepower
